Perform the following operations on the dataset:

● Drop Status id and status published column.

● Use a label encoder to encode the status type column.

● Standardize the data using min-max scalar

● Create a K means model for 2 clusters.

 From the above model, what is the inter-cluster variance of the model?

```python
##Let's delete the col which has the null valures
df.drop(columns=["Column1","Column2","Column3","Column4"],axis = 1, inplace = True)
```

```python
#● Drop Status id and status published column.
df.drop(columns = ["status_id", "status_published" ],axis = 1, inplace=True)
```

```python
##● Use a label encoder to encode the status type column.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["status_type"] = le.fit_transform(df["status_type"])
```

```python
#● Standardize the data using min-max scalar
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
```

```python
#● Create a K means model for 2 clusters.
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2,random_state=0)
kmeans.fit(df_scaled)
```

```python
#what is the inter-cluster variance of the model?
kmeans.inertia_
```

```python
# Get the inter-cluster variance (inertia) of the model
inter_cluster_variance = kmeans.inertia_
print(f"The inter-cluster variance of the model is: {inter_cluster_variance}")
```
