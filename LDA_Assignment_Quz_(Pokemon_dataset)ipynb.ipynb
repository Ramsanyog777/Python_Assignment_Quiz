25) On what basis the features are constructed in LDA

minimizes the differences between classes and maximizes the variations within each class.

maximizes the differences between classes and maximizes the variations within each class.

Ans

In Linear Discriminant Analysis (LDA), the features are constructed on the basis of maximizing the differences between classes and minimizing the variations within each class.

LDA seeks to find linear combinations of the original features that best separate the different classes. It achieves this by maximizing the ratio of between-class variance to within-class variance.



26) Use Pokemon.csv for the question below, remove â€˜nameâ€™ column and do not perform any data cleaning or data preprocessing before answering the question. Which of the dominant independent features that significantly influences the separability of the class fighting in LDA preprojection post performing below operations? Store column type (Target variable) as y Store Remaining columns as X Normalize the X and get X_norm using normalize in sklearn.preprocessing Apply LDA with n_components as 5 Display LDA coefficients


```python
from sklearn.preprocessing import normalize
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import pandas as pd

# Assuming 'type' is the target variable and the remaining columns are features
X = df.drop('type', axis=1)
y = df['type']

# Normalize the features
X_norm = normalize(X)

# Convert normalized features back to a DataFrame for easier analysis with column names
X_norm_df = pd.DataFrame(X_norm, columns=X.columns)

display(X_norm_df.head())
```

```python
# Apply LDA with n_components as 5
lda = LinearDiscriminantAnalysis(n_components=5)

# Fit LDA to the normalized data
lda.fit(X_norm_df, y)

# Display LDA coefficients
# The coefficients are stored in lda.coef_
# To understand the influence of each feature, we can look at the absolute values of the coefficients
# for each component and each class. However, the question asks about the influence on the
# separability of the 'fighting' class specifically. We can examine the coefficients associated
# with the 'fighting' class in the LDA model.

# Get the index of the 'fighting' class
fighting_class_index = lda.classes_.tolist().index('fighting')

print(f"LDA Coefficients for 'fighting' class:")
# lda.coef_ has shape (n_classes, n_features)
# We are interested in the row corresponding to the 'fighting' class
fighting_coefficients = lda.coef_[fighting_class_index, :]

# Create a pandas Series for better visualization with feature names
fighting_coefficients_series = pd.Series(fighting_coefficients, index=X.columns)

# Display the coefficients, sorted by absolute value to see dominant features
print(fighting_coefficients_series.abs().sort_values(ascending=False))
```
